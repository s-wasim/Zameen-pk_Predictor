##### BoxPlot
Creating box plots help analyse what relation exist within the ranges of the data. This would help identify outliers in our data so we can remove them for better prediction

```python
def boxplot(dataframe, col, adjusted_y = 0, figSize = (5,3)):
    medianprops = {'color': 'red'}
    boxprops = {'color': 'blue', 'linestyle': '-'}
    whiskerprops = {'color': 'green', 'linestyle': '--'}

    b_plot = df.boxplot(column = col, showfliers=False, medianprops=medianprops, boxprops=boxprops, whiskerprops=whiskerprops)
    b_plot.plot()


    median = df[col].median()
    plt.text(1, median + adjusted_y, median, horizontalalignment='center', color='red', fontsize = 8)

    # label the first quartile (Q1) value
    q1 = df[col].quantile(0.25)
    plt.text(1, q1 + adjusted_y, q1, horizontalalignment='center', color='black', fontsize=8)

    # label the third quartile (Q3) value
    q3 = df[col].quantile(0.75)
    plt.text(1, q3 + adjusted_y, q3,color='black', horizontalalignment='center', fontsize=8)

    #find the IQR
    iqr = q3 - q1
    lower_whisker = max(df.Price.min(), q1 - 1.5 * iqr)
    upper_whisker = min(df.Price.max(), q3 + 1.5 * iqr)

    plt.text(x=1, y= upper_whisker + adjusted_y, s = str(upper_whisker) + "(upper whisker)", horizontalalignment = 'center', fontsize = 8)
    plt.text(x=1, y= lower_whisker + adjusted_y, s = str(lower_whisker) + " (lower whisker)", horizontalalignment = 'center', fontsize = 8)

    plt.figure(figsize= figSize)
    plt.ylabel("Price in Crores/10 Million")
    plt.show() 
```

##### Histogram
A histogram would allow to analyse the ranges of data well which would assist in the analysis ahead.

```python
def histogram(data,_bins,adjusted_x_mul=1, adjusted_x_add=0, adjusted_y=0, x_label='', _title = ''):
    n, bins, patches =plt.hist(data,bins=_bins, rwidth=0.8)

    vals = n.tolist()
    plt.title=_title
    plt.xlabel(x_label)
    
    for i in range(0,len(vals)):
        plt.text(x = i*adjusted_x_mul + adjusted_x_add,y = vals[i]+adjusted_y,s= vals[i], color='black', fontsize=8)
    plt.show()
```

##### L192367 - Laiba Gohar
##### Bar graph
A bargraph is used to visualise the frequency of different variables.

```python
def bar_graph(data,y_adjusted,title):
    
    types = data.value_counts() 

    types.plot.bar()
    plt.ylabel("Count")
    plt.title = title
    
    for i in range(len(types)):
        plt.text(i - 0.25, types[i] + y_adjusted, types[i],color='black', fontsize=8)
    plt.show()
```
